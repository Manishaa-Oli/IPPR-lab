##Lab 5: Adding Gaussian noise to an image

###Theory:
Gaussian noise (also called normal noise) is a type of random noise whose pixel values follow a Gaussian (normal) distribution. It commonly occurs during image acquisition due to sensor noise, electronic circuit noise, or poor lighting conditions.

###Write a program to find erosion of an image


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/girl.jpg', 0)
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(img, kernel, iterations=1)

plt.figure(figsize=(12, 6))  #increases display size

plt.subplot(121)
plt.imshow(img, cmap="gray")
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(erosion, cmap="gray")
plt.title('Erosion')
plt.xticks([]), plt.yticks([])

plt.show()
```

###Dilation of an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/girl.jpg', 0)
kernel = np.ones((3,3), np.uint8)
dilation = cv2.dilate(img, kernel, iterations=1)

plt.figure(figsize=(12, 6))

plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dilation, cmap="gray"), plt.title('dilation')
plt.xticks([]), plt.yticks([])
plt.show()
```

###Opening of an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/girl.jpg', 0)
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(opening), plt.title('Opening')
plt.xticks([]), plt.yticks([])
plt.show()
```

###Closing of an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/girl.jpg', 0)
kernel = np.ones((3, 3), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(closing), plt.title('Closing')
plt.xticks([]), plt.yticks([])
plt.show()
```

###Morphological gradient of an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/girl.jpg', 0)
kernel = np.ones((3, 3), np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gradient), plt.title('Gradient')
plt.xticks([]), plt.yticks([])
plt.show()
```
